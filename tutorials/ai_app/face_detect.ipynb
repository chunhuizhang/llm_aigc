```python
import face_detection
import matplotlib.pyplot as plt
```

```python
face_detection.available_detectors
```

```python
detector = face_detection.build_detector(
  "DSFDDetector", confidence_threshold=.5, nms_iou_threshold=.3)
```

```python
import cv2
```

```python
im = cv2.imread('./imgs/0_Parade_marchingband_1_765.jpg')[:, :, ::-1]
detections = detector.detect(im)

# 
plt.imshow(im)
ax = plt.gca()

# 
for detection in detections:
    x, y, w, h, confidence = detection
    rect = plt.Rectangle((x, y), w-x, h-y, fill=False, color='red')
    ax.add_patch(rect)
    # ax.text(x, y, f'{confidence:.2f}', bbox=dict(facecolor='white', alpha=0.5))
```
