$$
\begin{split}
&\sigma(x) = \frac{1}{1 + e^{-x}}\\
&\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))
\end{split}
$$

- 梯度集中在 0 附近，`[-3, 3]`

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set_theme()

# Define the sigmoid function and its gradient
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_grad(x):
    s = sigmoid(x)
    return s * (1 - s)
```

```python
# Generate data points
x = np.linspace(-10, 10, 500)
y_sigmoid = sigmoid(x)
y_grad = sigmoid_grad(x)

# Create a plot for sigmoid and its gradient
plt.figure(figsize=(5, 3))
plt.plot(x, y_sigmoid, label='Sigmoid Function', color='blue')
plt.plot(x, y_grad, label='Gradient of Sigmoid', color='orange')
plt.title('Sigmoid Function and Its Gradient')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
